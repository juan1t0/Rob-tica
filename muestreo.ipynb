# Muestreo PDF

```
import matplotlib.pyplot as plt
import numpy as np

from math import sqrt
import random
```

```
def normalize( arraynp ):
    mean = np.mean(arraynp,0)
    standard = np.std(arraynp,0)
    for i in range(0,arraynp.shape[0]):
        arraynp[i] = np.true_divide((arraynp[i] - mean), standard)
    return arraynp
```

## Función para una distribución normal

```
def norm(b=1.0, r=12):
  nrm = np.random.uniform(low=-b, high=b, size=(r,))
  nrm = np.sum(nrm)
  return nrm / 2.0
```

## Función para una distribución triangular

```
raiz_6_entre2 = sqrt(6.0)/2
raiz_6_entre2
```

```
def triangular(b=1):
  r = random.uniform(-b,b) + random.uniform(-b,b)
  return r * raiz_6_entre2
```

## Ejercicios de muestreo

```
def muestreoPDF(npoints=100000, funcion=None, step=0.05, r=12):
  vals = np.zeros(npoints)
  for i in range(npoints):
    if funcion == 'N':
      vals[i]=norm(r=r)
    elif funcion == 'T':
      vals[i]=triangular()
    else:
      raise Exception("no function")
  x = np.arange(np.rint(np.min(vals)),
                np.rint(np.max(vals)), step)
  y = np.zeros(x.shape)
  
  for v in vals:
    for id,i in enumerate(x):
      if v >= i and v < (i+0.05):
        y[id] += 1
        break
  y = normalize(y)
  return x,y
```

### Ejemplos con 10,000 muestras

```
x,y = muestreoPDF(npoints=10000, funcion='N')

plt.fill_between(x,np.min(y),y, facecolor='skyblue')
plt.scatter(x,y,5)
plt.title('Distribución normal: 10,000 muestras')
plt.show()
```

```
x,y = muestreoPDF(npoints=10000, funcion='T')

plt.fill_between(x,np.min(y),y, facecolor='skyblue')
plt.scatter(x,y,5)
plt.title('Distribución triangular: 10,000 muestras')
plt.show()
```
